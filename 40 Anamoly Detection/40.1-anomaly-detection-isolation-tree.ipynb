```python
# 0 and 1 are some test parameters based on which the disease is identified. Not everyone will have disease.

import pandas as pd
df=pd.read_csv('healthcare.csv')
df.head()
```

```python
from matplotlib import pyplot as plt
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest

islnfrst = IsolationForest(contamination=0.2)
islnfrst.fit(df)
prediction = islnfrst.predict(df)
prediction 
```

```python
import numpy as np

index = np.where(prediction < 0)

index
```

```python
# collect dataframe values into an array
from matplotlib import pyplot as plt
x = df.values

index = np.where(prediction < 0)

plt.scatter(df.iloc[:,0], df.iloc[:,1])



plt.scatter(x[index,0], x[index,1],edgecolors='r')


```
