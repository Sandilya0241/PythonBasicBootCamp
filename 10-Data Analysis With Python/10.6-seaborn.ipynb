### Data Visualization With Seaborn
* Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics.
* Seaborn helps in creating complex visualizations with just a few lines of code.


```python
import seaborn as sns
## Basic plotting with Seaborn

tips_ds=sns.load_dataset("tips")
print(tips_ds)

## Create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x="total_bill", y="tip",data=tips_ds)
plt.title("Total Bill Vs Tip")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips_ds=sns.load_dataset("tips")
sns.lineplot(x="size", y="total_bill",data=tips_ds)
plt.title("Line plot of Total Bill by Size")
plt.show()
```

```python
## Categorical plots
import seaborn as sns
import matplotlib.pyplot as plt

tips_ds=sns.load_dataset("tips")
sns.barplot(x="day",y="total_bill",data=tips_ds)
plt.title("Bar plot of Total Bill by Day")
plt.show()
```

```python
## Box plot
import seaborn as sns
import matplotlib.pyplot as plt

tips_ds=sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=tips_ds)
plt.title("Box plot of Total Bill by Day")
plt.show()
```

```python
## To see distribution we can use Violoin plot
import seaborn as sns
import matplotlib.pyplot as plt

tips_ds=sns.load_dataset("tips")
sns.violinplot(x="day",y="total_bill",data=tips_ds)
plt.title("Violin plot of Total Bill by Day")
plt.show()
```

```python
## Histogram plot
import seaborn as sns
import matplotlib.pyplot as plt

tips_ds=sns.load_dataset("tips")
sns.histplot(tips_ds['total_bill'],bins=10,kde=True)
plt.title("Histogram plot of Total Bills")
plt.show()
```

```python
## KDE plot
import seaborn as sns
import matplotlib.pyplot as plt

tips_ds=sns.load_dataset("tips")
sns.kdeplot(tips_ds['total_bill'],fill=True)
plt.title("KDE plot of Total Bills")
plt.show()
```

```python
## Pairplot plot
import seaborn as sns
import matplotlib.pyplot as plt

tips_ds=sns.load_dataset("tips")
sns.pairplot(tips_ds)
```

```python
## KDE plot
import seaborn as sns
import matplotlib.pyplot as plt

tips_ds=sns.load_dataset("tips")
sns.kdeplot(tips_ds['total_bill'],fill=True)
plt.title("KDE plot of Total Bills")
plt.show()
```

```python
## Correlation plot/ Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

tips_ds = sns.load_dataset("tips")
corr = tips_ds[['total_bill','tip','size']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
## Customize a plot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sales_df = pd.read_csv("1000 Sales Records.csv")
# print(sales_df)
# Plot 'Units sold' against 'Item Type'
plt.figure(figsize=(20,6))
sns.barplot(x='Item Type',y='Total Revenue',data=sales_df,estimator=sum) 
plt.title('Total Sales By ItemType')
plt.xlabel('Item Type')
plt.ylabel('Total Revenue')
plt.show()
```
