### **HISTOGRAM**

Ages = {25, 24, 25, 30, 34, 36, 40, 50, 60, 75, 80}



* Histogram helps us to find the frequency of a specific distribution. 

* Using Histogram we can derive `Probability Distribution Function` or `Probability Density Function`.



* Smoothened Histogram is called `Probability Distribution Function` or `Probability Density Function`. 

* With the help of `Kernel Density Estimator` we can draw `Probability Distribution Function` or `Probability Density Function`.

* **NOTE: EVEN THOUGH THERE ARE LOT OF OVERLAPPING POINTS IN DATA, HISTOGRAM ALWAYS CONSIDER FREQUENCY OF ELEMENT THAT ARE RANGING BETWEEN THE BINS.**
