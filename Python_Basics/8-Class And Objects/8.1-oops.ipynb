```python
## Instance Variable
class Dog:
    # constructor
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed
    

dogInstance = Dog('Jimmy', 'Boxer')
print(dogInstance)
print(dogInstance.name)
print(dogInstance.breed)
```

```python
## Instance Method

class Dog:
    
    ## Constructor
    def __init__(self,name, age):
        self.name = name
        self.age = age
    
    ## Instance method
    def bark(self):
        print(f'{self.name} says woof')


dog1 = Dog("buddy", 3)
dog1.bark()

dog2 = Dog('Lucy', 4)
dog2.bark()
```

```python
## Modelling a bank account

## Define a class for bank account

class Account:
    
    def __init__(self, owner, balance = 0.0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f'Amount deposited. New balance is {self.balance}')

    def withdraw(self, amount):
        if amount > self.balance:
            print(f'Insufficient funds!!!')
        else:
            self.balance -= amount
            print(f'{amount} was with drawn. New balance is {self.balance}')

    def get_balance(self):
        return self.balance
    


## Create account
sandy_account = Account('Sandy', 5000)
print(sandy_account.balance)

sandy_account.deposit(1000)

sandy_account.withdraw(2000)

sandy_account.withdraw(5000)

sandy_account.get_balance()

    
```

**Instance Variables vs Class Variables in Python**

__Instance Variables:__

Variables that are unique to each object. They are created inside methods (typically __init__()).

__Class Variables (Static Variables):__ 

Variables shared among all instances of a class. They are defined within the class, outside any methods.

The classmethod() is an inbuilt function in Python, which returns a class method for a given function. This means that classmethod() is a built-in Python function that transforms a regular method into a class method. When a method is defined using the @classmethod decorator (which internally calls classmethod()), the method is bound to the class and not to an instance of the class. As a result, the method receives the class (cls) as its first argument, rather than an instance (self).

classmethod() in Python Syntax
class MyClass:

    @classmethod

    def class_method(cls, *args, **kwargs):

        # Method implementation
