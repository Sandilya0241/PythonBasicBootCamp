### Pandas-DataFrame And Series
* Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning.
* It provides two primary data structure:
    - Series
    - DataFrame
* A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogenous tabular data structure with labelled axes (rows and columns).

```python
import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
## Create a series from dictionary elements
data = {'a':1,'b':2,'c':3}
series_data=pd.Series(data)
print(series_data)
```

```python
data=[10,20,30]
indexes=['a','b','c']
pd.Series(data,index=indexes)
```

```python
## DataFrame
## Create a DataFrame from a dictionary of list
data={
    'name':['Sandy','John','Jack'],
    'Age':[36,37,38],
    'City':['Dubai','Bangalore','Hyderabad']
}
data_frame = pd.DataFrame(data)
print(data_frame)
print(type(data_frame))
```

```python
import numpy as np
np.array(data_frame)
```
