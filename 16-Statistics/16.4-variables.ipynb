### **VARIABLE**



* Defnition: Variable is a property that can take up any value.

    Ex: Age = 25 is a variable

    But Ages = [12, 24, 25,14] is not a variable

* **Different Type Of Variable**



#### **QUANTITATIVE VARIABLE**
    
- **DISCRETE QUANTITATIVE VARIABLE** : A variable that can be specified as a whole number. For example, Person A has 2 Children/ 3 Childern. But we cannot say 2.5 Childern.


- **CONTINUOUS QUANTITATIVE VARIABLE** : Example is height. Height can be 168.5 cms/ 170 cms/ 151.2 cms.

#### **QUALITATIVE/ CATEGORICAL VARIABLE**

- Examples are 
    1. Gender - Male/ Female/ Other
    2. Color - Red/ Green/...
