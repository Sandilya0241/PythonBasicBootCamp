### **COVARIANCE AND CORRELATION**

* **`Covariance`** and **`Correlation`** are two statistical measure used to determine the relationship between two variables.
* Both are used to understand how changes in one variable are associated with changes in another variable.

<br>

**COVARIANCE** :
* Covariance is a measure of how much two random variables changes together.
* If the variables tend to increase or decrease together, the Covariance is **`POSITIVE`**.
* If the one tends to increase when other decreases, the Covariance is **`NEGATIVE`**.
* Quantify the relation between variables

    If X INC Y INC

    IF X DEC Y INC

    If X INC Y DEC

    IF X DEC Y DEC

    EX:

    


    **POSITIVE COVARIANCE EXAMPLE**

    X INC -> Y INC

    X DEC -> Y DEC

         


    **NEGATIVE COVARIANCE EXAMPLE**

    X INC -> Y DEC

    X DEC -> Y INC


          


### **COVARIANCE FORMULA**



COV(X,X) is Nothing but Variance



#### Let us check above formula with some example:

 <u>Example 1:</u> 



xbar = (2+3+4+5+6)/5 = 4

ybar = (50+60+70+80+90)/5 = 70

Cov(X,Y) = ((2-4)(50-70) + (3-4)(60-70) + (4-4)(70-70) + (5-4)(80-70) + (6-4)(90-70))/(5-1) = (40 + 10 + 0 + 10 + 40)/4 = 25

 <u>Example 2:</u>



xbar = (7 + 6 + 5)/3 = 6

ybar = (10 + 12 + 14)/3 = 12

Cov(X,Y) = ((7-6)(10-12) + (6-6)(12-12) + (5-6)(14-12) )/(3-1) = (-2 + 0 + (-2))/4 = -1


#### **Covariance Advantage**:
* Quantify the relationship between X and Y.

#### **Covariance Disvantage**:
* Covariance doesn't have a specific unit value (`-INFINITY` to `+INFINITY`).
* If we have A, B, C random variables. 

    COV(A,B) = 20

    COV(A,c) = 30

    COV(B,C) = 40

    As there is no limit, we can never say COV(B,C) > COV(A,B).
