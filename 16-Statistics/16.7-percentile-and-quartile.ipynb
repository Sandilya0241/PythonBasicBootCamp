### **PERCENTILE AND QUARTILE**

* **PERCENTILE** : A Percentile is a value below which certain percentage of observations lie.

    * **Percentile(%) of X = (Count of values below X) / n(Count of all values) * 100**

    - Let us take this distribution {2,2,3,4,5,5,6,7,8,8,8,9,9,10}
    - Now calculate % of 9:
        - 11/14 * 100 = 78.57% (This means 78% of numbers are less than 9)
    - Now calculate 25% of this distribution:
        - **Value = Percentile(%) * (n+1) / 100**
        - 25/100 * 15 = 3.75 ~= 3.5
<br>
* **QUARTILE** : A Percentile is a value below which certain percentage of observations lie.
    * 25% = First Quartile
    * 50% = Second Quartile
    * 75% = Third Quartile
    * Every 25% is a quartile.

### **5 NUMBER SUMMARY**: This can be used when we are performing some tasks in Machine Learning like Feature Selection, How to remove Outliers.

1) **`MINIMUM`**
2) **`1ST QUARTILE (25%) (Q1)`**
3) **`MEDIAN`**
4) **`3RD QUARTILE (75%) (Q3)`**
5) **`MAXIMUM`**

* STEP 1: DEFINE LOWER FENCE VALUE.
    * **LOWER FENCE = Q1 - 1.5 * (IQR)**
* STEP 2: DEFINE HIGHER FENCE VALUE.
    * **HIGHER FENCE = Q3 + 1.5 * (IQR)**
* **IQR (INTER QUARTILE RANGE) = Q3-Q1**
* **`VALUES LESS THAN LOWER FENCE AND GREATER THAN HIGHER FENCE ARE CONSIDERED AS OUTLIERS`**
* Now, let us do this for this sample {1,2,2,2,3,3,4,5,5,5,6,6,6,6,7,8,8,9,27}

* `Q1` = **PERCENTILE(%) * (n+1) / 100**
    
    = 25 / 100 * (19 + 1) = 5th Position.
    
    = Number at 5th Position is 3

* `Q3` = **PERCENTILE(%) * (n+1) / 100**
    
    = 75 / 100 * (19 + 1) = 3 / 4 * 20 = 15th Position.
    
    = Number at 15th Position is 7

* `IQR` = **Q3 - Q1**
    
    = 7 - 3

    = 4

* `LOWER FENCE` = **Q1 - 1.5 * (IQR)**
    
    = 3 - 1.5 * 4

    = -3

* `HIGHER FENCE` = **Q3 + 1.5 * (IQR)**
    
    = 7 + 1.5 * 4

    = 13

* `DATA RANGE` = **[-3, 13]**

* So, 27 is an outlier. After removing outlier, distribution is {1,2,2,2,3,3,4,5,5,5,6,6,6,6,7,8,8,9}

* **`MINIMUM` = 1**

* **`1ST QUARTILE (25%) (Q1)` = 3**

* **`MEDIAN` = (5+5) / 2 = 5**

* **`3RD QUARTILE (75%) (Q3)` = 7**

* **`MAXIMUM` = 9**

* **BASED ON ABOVE VALUES WE CAN PLOT `BOX PLOT`. THIS CAN HELP US SEE/ VISUALIZE OUTLIERS. THERE IS A SIMILAR PLOT CALLED `WHISKER PLOT`**
