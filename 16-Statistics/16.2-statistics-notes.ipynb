* **Median**

    -   Let us try to understand with earlier Ages example. Let us say there is an outlier(will be explained more further in detail)/ big value got added in that list like {1,3,4,5,`100`}. With this number the mean changes from `3.25` to `22.6`. There are ways to remove those outliers. In order to overcome the impact of this particular big number, what we do is called as `Median`.

        -   Step 1: Sort the numbers. 

            - {4,3,5,1,100} -> {1,3,4,5,100}

        -   Step 2: Pick the center element if the count of elements are `ODD`. If the count of elements is even, we have to calculate the average of center 2 elements.

            - {1,3,4,5,100} -> Median = 4
            - {1,3,4,5,100,200} -> Median = (4+5)/2 = 4.5

* **Mode**

    -   We select the element which has maximum frequency.

        -   {4,3,2,1,14,4,5,2,100} - Mode is 4.
